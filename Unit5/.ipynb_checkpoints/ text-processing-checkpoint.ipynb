### Data Cleaning

#### Import Wuthering Heights Text and Needed Libraries

```python
import re
import numpy as np
import random

with open('wutheringheights.txt', 'r', encoding="utf8") as file:
    og_txt = file.read().replace('\n', ' ').lower()
og_txt
```

#### Remove All Non-Letter Characters

```python
txt = re.sub("[^a-zA-Z]+", " ",og_txt)
txt
```

#### Turn Text into a List of Words

```python
words = txt.split()
```

#### Remove Chapter Titles and Other Messy Words

```python
messy = ['chapter','m',' ','s','ll']
words = [x for x in words if x not in messy ]
```

#### Remove Roman Numerals

```python
words.remove('i')
num = ['ii','iii','iv','v','vi','vii','viii','ix','x','xi','xii','xiii','xiv','xv','xvi','xvii','xviii','xix',
       'xx','xxi','xxii','xxiii','xxiv','xxv','xxvi','xxvii','xxviii','xxix','xxx','xxxi','xxxii','xxxiii','xxxiv']
words = [x for x in words if x not in num ]
words
```

#### Turn into NumPy Array

```python
words = np.asarray(words)
```

### Creating A Dictionary

#### Set of Unique Words

```python
unique_words = set(words)
unique_words.remove('earth')
```

#### Next Words Dictionary

```python
def getNxt(word):
    indices = np.where(words==word)
    indices = indices[0]
    nxt_words = [words[index+1] for index in indices]
    return nxt_words
```

```python
pred_dict = {word:getNxt(word) for word in unique_words}
```

### III. Sentence Generation with RECURSION!

```python
sen = []

def makeSen(word, senLen):
    global sen
    sen.append(word)
    if len(sen) < senLen:
        options = pred_dict.get(word) # get following words
        nxt = random.choice(options) #choose random following word (works better than most common word)
        makeSen(nxt, senLen)
    else:
        sen = ' '.join(sen)
        return sen
```

### Try it Out!

```python
print('Write a word to start your sentence!')
start_word = input()
makeSen(start_word, 26)
```
