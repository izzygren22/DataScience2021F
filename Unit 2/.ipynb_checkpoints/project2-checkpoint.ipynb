# College Majors: What The Data Tells Us

## What is the correlation betweeen women majority jobs and low salary?

#### Ingest Data From Local Comma Seperate Value File

```python
import pandas as pd
majors = pd.read_csv('college_majors.csv')
pd.set_option('display.max_columns', None)
majors
```

#### Checking for complete data - are there any missing values?

```python
majors[majors.isnull().any(axis=1)] #datatofish
majors.dropna()
```

#### Describe the Data

```python
majors.describe()
```

### Before We Dig Deeper

Average Share Women

```python
avgShrWmn = majors.ShareWomen.mean()
avgShrWmn
```

Average Unemployment Rate

```python
avgUnemp = majors.Unemployment_rate.mean()
avgUnemp
```

NOTICE THEIR IS A .64 CORRALATION BETWEEN SALARY RANK AND SHARE OF WOMEN, LET'S EXPLORE MORE...

```python
majors.corr()
```

### Majority Women Fields

Series of Fields' Share of Women

```python
ShrWmnSr = majors.groupby('Major_category').ShareWomen.mean()
ShrWmnSr
```

Switch from Series to DataFrame

```python
ShrWmn = ShrWmnSr.to_frame()
ShrWmn
```

Sort by Majority Women

```python
ShrWmn = ShrWmn.sort_values(by='ShareWomen', ascending=False)
ShrWmn = ShrWmn.reset_index()
ShrWmn
```

### Highest Salary Fields 

Series of Fields' average salary rank

```python
SalSr = majors.groupby('Major_category').Rank.mean()
SalSr
```

Switch Series to DataFrame

```python
SalRank = SalSr.to_frame()
SalRank
```

Sort by Lowing Average Salary

```python
SalRank = SalRank.sort_values(by='Rank', ascending=False)
SalRank = SalRank.reset_index()
SalRank
```

### Let's Compare Our Graphs

Delete Pure Data (Rank and Share Women) from both Dataframes

```python
CopyShrWmn = ShrWmn.drop(columns='ShareWomen')
CopyShrWmn
```

```python
CopySalRank = SalRank.drop(columns='Rank')
CopySalRank
```

### Compare Share Women Fields Database to the Lowest Paid Fields Database
- The NaNs Denote where the Majors are the same (3 is Education and 15 is Enginnering)

```python
CopyShrWmn.compare(CopySalRank, keep_shape=True)
```
