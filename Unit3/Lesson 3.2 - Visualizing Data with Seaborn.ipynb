# Data Science Lesson - Visualizing Data with Seaborn
---
Let's import pandas and seaborn, and then use our small "people" data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_file_path = 'people.csv'
people = pd.read_csv(data_file_path)
people
```

Now let's create a linear model of friends to gpa:

```python
sns.lmplot(x='friends', y='gpa', data=people)
plt.show()
```

The default seaborn linear model plot (lmplot) fits a regression line and indicates 95% confidence interval.

## Bar Charts

Let's load the wine review data to find out how various countries stack up against each other for the finest wine.

```python
wr = pd.read_csv('winemag-data_first150k.csv', index_col=0)
best = wr.loc[wr.points == wr.points.max()]
best
```

```python
best_counts = best.value_counts('country')
best_counts
```

```python
sns.barplot(x = best_counts.index, y = best_counts)
plt.show()
```

Now, if we want a bit more control over the chart, we need to use matplotib, like this:

```python
sns.barplot(x = best_counts.index, y = best_counts) # same as before

plt.title('Number of "best" wines by country', fontsize = 14)
plt.grid(axis='y', zorder=0)
plt.show()
```
